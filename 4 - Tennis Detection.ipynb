# Part 2: Detection using Thresholding


Step 1: Read in the image "tennis.jpg" image and display it


```python
# Load a color image
img = cv2.imread('images/tennis.jpg')
img.shape
img = cv2.resize(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), dsize=(320,180))
plt.imshow(img)
plt.show()
```

Step 2: Find threshold to use for tennis ball by changing the values for lower and upper. Hint: order is "R,G,B" and numbers are between 0, 255 for each!

```python
### START OF YOUR CODE:
lower = np.array([20,45,60])
upper = np.array([70,100,200])
### END OF YOUR CODE

mask = cv2.inRange(img, lower, upper)
res = cv2.bitwise_and(img, img, mask= mask)
plt.imshow(res)
```
